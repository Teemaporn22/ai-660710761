# เติมค่า null ด้วยค่าเฉลี่ย

```python
df_mean = df.copy()
```

```python
# เติมค่าอายุด้วยค่าเฉลี่ย อายุก็จะไม่มีค่า NaN
# เหมาะกับข้อมูลตัวเลข
mean_age = df_mean['อายุ'].mean()
mean_grade = df_mean['GPA'].mean()
mean_height = df_mean['ส่วนสูง_cm'].mean()
mean_weight = df_mean['น้ำหนัก_kg'].mean()
mean_hour = df_mean['ชั่วโมงเรียน_ต่อสัปดาห์'].mean()

#แทนค่าที่มี NaN ด้วยค่าเฉลี่ย ด้วยฟังก์ชั่น fillna
df_mean['อายุ'].fillna(mean_age, inplace=True)
df_mean['GPA'].fillna(mean_grade,inplace=True)
df_mean['ส่วนสูง_cm'].fillna(mean_height, inplace=True)
df_mean['น้ำหนัก_kg'].fillna(mean_weight, inplace=True)
df_mean['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(mean_hour, inplace=True)
```

```python
print(f"\nค่าเฉลี่ยอายุ -> {mean_age:.1f} ปี")
print("หลังเติมค่า")
df_mean
```

```python
df_mean.isnull().sum()
```

```python
print("ค่า NaN ที่ยังเหลืออยู่")
df_mean.isnull().sum().sum()
```

```python
df_mode.isnull().sum()
```

# ทำให้ข้อมูลเป็นมาตรฐาน

```python
# ทำให้ชื่อคณะเป็นมาตรฐานเดียวกัน
# ข้อมูลบางข้อมูล เช่น คณะ มีการใส่ วิทย์,วิทยาศาสตร์,Sci ค่าไม่เหมือนกัน ดังนั้น จึงต้องทำให้เป็นมาตรฐานเดียวกัน
df_standard = df_unknown.copy()
```

```python
df_standard['คณะ']
```

```python
df_standard['เพศ']
```

```python
#สร้างฟังก์ชั่น

def standardize_gender(name):
    if name in ['ชาย','Male','M','ช']:
        return 'ชาย'
    elif name in ['หญิง','Female','F','ญ']:
        return 'หญิง'
    else:
        return name

def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์','Sci']: #วิทยาศาสตร์ - วิทย์ , Science ,Sci
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts','Liberal Arts']:#ศิลปศาสตร์ - Arts,ศิลป์ ,Liberal Arts
        return 'ศิลปศาสตร์'
    elif name in ['วิศวกรรมศาสตร์','วิศวกรรม','วิศวะ','Engineering','วิศวฯ']: #วิศวกรรมศาสตร์ - วิศวกรรม,วิศวะ ,Engineering ,วิศวฯ
        return 'วิศวกรรมศาสตร์'
    else:
        return name



#แทนที่ชื่อคณะให้เหมือนกัน
df_standard['เพศ'] = df_standard['เพศ'].apply(standardize_gender)
df_standard['คณะ'] = df_standard['คณะ'].apply(standardize_faculty)
df_standard
```

```python
#Age
Q1_Age = df['อายุ'].quantile(0.25)
Q3_Age = df['อายุ'].quantile(0.75)
IQR_Age = Q3_Age - Q1_Age
#ขอบเขตของ อายุ
lower_bound_age = Q1_Age - 1.5 * IQR_Age
upper_bound_age = Q3_Age + 1.5 * IQR_Age
print(lower_bound_age,upper_bound_age)

#gpa
lower_bound_gpa = 1
upper_bound_gpa = 4
print(lower_bound_gpa,upper_bound_gpa)
```

```python
#กรองค่าที่เกินขอบเขต ของอายุ และเกรด
df_no_outliers = df[(df['อายุ'] >= lower_bound_age) & (df['อายุ'] <= upper_bound_age ) & (df['GPA'] >= 0) & (df['GPA'] <= 4)]
df = df_no_outliers
df
```

```python
# สร้าง BMI จากส่วนสูงและน้ำหนัก
df_bmi = df_mode.copy()
```

```python
# สร้าง BMI
df_bmi['BMI'] = df_bmi['น้ำหนัก_kg'] / (df_bmi['ส่วนสูง_cm']/100)**2
df_bmi['BMI']
```
